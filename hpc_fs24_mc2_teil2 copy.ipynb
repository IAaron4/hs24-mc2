```python
import cupy as cp
import numpy as np
from numba import cuda, float32
import math
import time
from typing import Dict, List, Tuple
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Test matrices configurations (M, K, N) for multiplication of shape (M,K) * (K,N)
test_matrices = [
    (5120, 5120, 5120),     # ~26M elements
    (8192, 4096, 8192),     # ~34M elements
    (10240, 10240, 10240),  # ~105M elements
    (16384, 8192, 8192),    # ~134M elements
    (20480, 20480, 20480)   # ~419M elements
]

# Block configurations to test (threads_x, threads_y)
block_configs = [
    (8, 8),     # 64 threads
    (16, 16),   # 256 threads
    (32, 16),   # 512 threads
    (32, 32)    # 1024 threads
]

def cpu_matmul(A: np.ndarray, B: np.ndarray) -> np.ndarray:
    """CPU implementation for comparison"""
    return np.matmul(A, B)

def run_gpu_experiment(A: np.ndarray, B: np.ndarray, block_size: Tuple[int, int], 
                      num_warmup: int = 3, num_runs: int = 3) -> Dict:
    """Run GPU experiment with specific block size including warmup runs"""
    TPB_x, TPB_y = block_size
    M, K = A.shape
    K, N = B.shape
    
    # Copy to device
    A_gpu = cuda.to_device(A)
    B_gpu = cuda.to_device(B)
    C_gpu = cuda.device_array((M, N), dtype=np.float32)

    # Configure grid
    threadsperblock = block_size
    blockspergrid_x = math.ceil(M / TPB_x)
    blockspergrid_y = math.ceil(N / TPB_y)
    blockspergrid = (blockspergrid_x, blockspergrid_y)

    # Warmup runs
    print(f"Performing {num_warmup} warmup runs...")
    for _ in range(num_warmup):
        fast_matmul[blockspergrid, threadsperblock](A_gpu, B_gpu, C_gpu, TPB_x, TPB_y)
        cuda.synchronize()

    # Measurement runs
    times = []
    print(f"Performing {num_runs} measurement runs...")
    for i in range(num_runs):
        # Clear cache and synchronize
        cuda.synchronize()
        
        start = time.perf_counter()
        fast_matmul[blockspergrid, threadsperblock](A_gpu, B_gpu, C_gpu, TPB_x, TPB_y)
        cuda.synchronize()
        end = time.perf_counter()
        
        execution_time = end - start
        times.append(execution_time)
        print(f"Run {i+1}: {execution_time:.4f} seconds")

    return {
        'times': times,
        'mean': np.mean(times),
        'std': np.std(times),
        'min': np.min(times),
        'max': np.max(times)
    }

@cuda.jit
def fast_matmul(A, B, C, TPB_x, TPB_y):
    """CUDA kernel for matrix multiplication using shared memory"""
    # Define shared memory arrays
    sA = cuda.shared.array(shape=(32, 32), dtype=float32)  # Using maximum possible size
    sB = cuda.shared.array(shape=(32, 32), dtype=float32)

    x, y = cuda.grid(2)
    tx = cuda.threadIdx.x
    ty = cuda.threadIdx.y
    bpg = cuda.gridDim.x

    # Use actual block size for calculations
    TPB = min(TPB_x, TPB_y)

    tmp = float32(0.)
    for i in range(bpg):
        # Preload data into shared memory
        sA[ty, tx] = 0
        sB[ty, tx] = 0
        if y < A.shape[0] and (tx+i*TPB) < A.shape[1]:
            sA[ty, tx] = A[y, tx + i * TPB]
        if x < B.shape[1] and (ty+i*TPB) < B.shape[0]:
            sB[ty, tx] = B[ty + i * TPB, x]

        cuda.syncthreads()

        for j in range(TPB):
            tmp += sA[ty, j] * sB[j, tx]

        cuda.syncthreads()

    if y < C.shape[0] and x < C.shape[1]:
        C[y, x] = tmp

def run_experiments(save_results: bool = True):
    results = []
    
    for m1, m2, m3 in test_matrices:
        print(f"\nTesting matrix size: {m1}x{m2} * {m2}x{m3}")
        A = np.random.randn(m1, m2).astype(np.float32)
        B = np.random.randn(m2, m3).astype(np.float32)

        # CPU baseline
        print("\nRunning CPU baseline...")
        start = time.perf_counter()
        cpu_matmul(A, B)
        cpu_time = time.perf_counter() - start
        print(f"CPU Time: {cpu_time:.3f}s")

        # GPU experiments with different block sizes
        for block_size in block_configs:
            print(f"\nTesting block size: {block_size[0]}x{block_size[1]}")
            gpu_results = run_gpu_experiment(A, B, block_size)
            speedup = cpu_time / gpu_results['mean']
            
            results.append({
                'Matrix Size': f"{m1}x{m2}x{m3}",
                'Block Size': f"{block_size[0]}x{block_size[1]}",
                'CPU Time': cpu_time,
                'GPU Time Mean': gpu_results['mean'],
                'GPU Time Std': gpu_results['std'],
                'Speedup': speedup,
                'Elements': m1 * m2 * m3
            })
            
            print(f"Average GPU Time: {gpu_results['mean']:.4f}s (Speedup: {speedup:.2f}x)")
            print(f"Std Dev: {gpu_results['std']:.4f}s")

    results_df = pd.DataFrame(results)
    
    if save_results:
        results_df.to_csv('gpu_experiment_results.csv', index=False)
        
    return results_df

def visualize_results(df: pd.DataFrame):
    # Print detailed statistics
    print("\nDetailed Statistics:")
    print("=" * 80)
    stats = df.groupby(['Matrix Size', 'Block Size']).agg({
        'GPU Time Mean': ['mean', 'std'],
        'Speedup': ['mean', 'std']
    }).round(4)
    
    print(stats)
    
if __name__ == "__main__":
    results_df = run_experiments()
    visualize_results(results_df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Read each CSV file containing results for different block sizes
df_16 = pd.read_csv('./Work_Aaron/Exercise_2/gpu_experiment_results_16.csv')
df_24 = pd.read_csv('./Work_Aaron/Exercise_2/gpu_experiment_results_24.csv')
df_32 = pd.read_csv('./Work_Aaron/Exercise_2/gpu_experiment_results_32.csv')

# Add a column to identify the block configuration used for each set of results
df_16['Block_Config'] = '16x16'
df_24['Block_Config'] = '24x24'
df_32['Block_Config'] = '32x32'

# Combine all data into a single DataFrame for easier analysis
df_combined = pd.concat([df_16, df_24, df_32], ignore_index=True)

# Extract the matrix dimension (N from NxN) for proper sorting and plotting
# This converts strings like "1024x1024x1024" to just "1024" for the x-axis
df_combined['Matrix_Dim'] = df_combined['Matrix Size'].str.split('x').str[0].astype(int)


def create_speedup_comparison(df):
    """
    Create a visualization comparing speedup across different implementations and block sizes.
    
    This plot shows:
    1. Speedup relative to CPU for each implementation (Basic, Shared Memory, CuPy)
    2. Comparison across different block sizes (16x16, 24x24, 32x32)
    3. How speedup scales with matrix size
    
    The plot uses:
    - Log scales for both axes to better show relationships across different orders of magnitude
    - Different line styles for different block sizes
    - Different markers and colors for different implementations
    - Grid lines for easier reading of values
    
    Args:
        df (DataFrame): Combined performance data
        
    Returns:
        Figure: Matplotlib figure object containing the speedup comparison plot
    """
    plt.figure(figsize=(15, 8))
    
    # Define the implementations to plot and their labels
    implementations = [
        ('Basic GPU Speedup', 'Basic GPU'),
        ('Shared GPU Speedup', 'Shared Memory GPU'),
        ('CuPy GPU Speedup', 'CuPy')
    ]
    
    # Define distinct markers and colors for different implementations
    markers = ['o', 's', '^']  # Circle, Square, Triangle
    colors = ['#1f77b4', '#2ca02c', '#ff7f0e']  # Blue, Green, Orange
    
    # Create lines for each implementation and block size combination
    for (col, label), marker, color in zip(implementations, markers, colors):
        # Use different line styles for different block sizes
        for block_size, linestyle in zip(['16x16', '24x24', '32x32'], 
                                       ['-', '--', ':']):  # Solid, Dashed, Dotted
            data = df[df['Block_Config'] == block_size]
            plt.plot(data['Matrix_Dim'], data[col], 
                    label=f'{label} ({block_size})',
                    marker=marker, linestyle=linestyle,
                    color=color)

    # Set log scales for both axes as speedup and matrix size vary over orders of magnitude
    plt.xscale('log')
    plt.yscale('log')
    
    # Add grid for easier reading of values
    plt.grid(True, which="both", ls="-", alpha=0.2)
    
    # Label axes and add title
    plt.xlabel('Matrix Dimension (N for NxN)')
    plt.ylabel('Speedup vs CPU (log scale)')
    plt.title('GPU Implementation Speedup Comparison\nAcross Different Block Sizes and Matrix Dimensions')
    
    # Add legend with good placement
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    
    return plt.gcf()

def create_error_comparison(df):
    """
    Create a bar plot comparing maximum numerical errors across implementations.
    
    This plot shows:
    1. Maximum difference from CPU results for each implementation
    2. How errors vary with matrix size
    3. Comparison of numerical stability across implementations
    
    The plot uses:
    - Grouped bars for easy comparison between implementations
    - Log scale for error values as they can vary by orders of magnitude
    - Different colors for different implementations
    
    Args:
        df (DataFrame): Combined performance data
        
    Returns:
        Figure: Matplotlib figure object containing the error comparison plot
    """
    plt.figure(figsize=(15, 8))
    
    # Get unique matrix dimensions and prepare x-axis
    matrix_dims = sorted(df['Matrix_Dim'].unique())
    x = np.arange(len(matrix_dims))
    width = 0.25  # Width of bars
    
    # Define the error measurements to compare
    error_cols = [
        ('Basic GPU Max Diff', 'Basic GPU'),
        ('Shared GPU Max Diff', 'Shared Memory GPU'),
        ('CuPy GPU Max Diff', 'CuPy')
    ]
    
    # Create grouped bars for each implementation
    for i, (col, label) in enumerate(error_cols):
        # Calculate mean error across block sizes for each matrix dimension
        means = [df[df['Matrix_Dim'] == dim][col].mean() for dim in matrix_dims]
        plt.bar(x + i*width, means, width, label=label)

    # Use log scale for error values as they can vary by orders of magnitude
    plt.yscale('log')
    
    # Label axes and add title
    plt.xlabel('Matrix Dimension')
    plt.ylabel('Maximum Error (log scale)')
    plt.title('Maximum Error Comparison\nAcross Different Implementations and Matrix Sizes')
    
    # Create x-tick labels showing matrix dimensions
    plt.xticks(x + width, [f'{dim}x{dim}' for dim in matrix_dims], rotation=45)
    plt.legend()
    plt.grid(True, which="both", ls="-", alpha=0.2)
    plt.tight_layout()
    
    return plt.gcf()

def create_time_comparison(df):
    """
    Create a visualization comparing execution times across different implementations.
    
    This plot shows:
    1. Absolute execution time for each implementation
    2. How execution time scales with matrix size
    3. Performance differences between block sizes
    
    The plot uses:
    - Log scales for both axes as both time and matrix size vary over orders of magnitude
    - Different line styles for different block sizes
    - Different markers and colors for different implementations
    
    Args:
        df (DataFrame): Combined performance data
        
    Returns:
        Figure: Matplotlib figure object containing the time comparison plot
    """
    plt.figure(figsize=(15, 8))
    
    # Define the implementations to compare
    implementations = [
        ('Basic GPU Time', 'Basic GPU'),
        ('Shared GPU Time', 'Shared Memory GPU'),
        ('CuPy GPU Time', 'CuPy')
    ]
    
    # Define distinct markers and colors
    markers = ['o', 's', '^']
    colors = ['#1f77b4', '#2ca02c', '#ff7f0e']
    
    # Create lines for each implementation and block size
    for (col, label), marker, color in zip(implementations, markers, colors):
        for block_size, linestyle in zip(['16x16', '24x24', '32x32'], 
                                       ['-', '--', ':']):
            data = df[df['Block_Config'] == block_size]
            plt.plot(data['Matrix_Dim'], data[col], 
                    label=f'{label} ({block_size})',
                    marker=marker, linestyle=linestyle,
                    color=color)

    # Set log scales as both time and matrix size vary over orders of magnitude
    plt.xscale('log')
    plt.yscale('log')
    plt.grid(True, which="both", ls="-", alpha=0.2)
    
    # Label axes and add title
    plt.xlabel('Matrix Dimension (N for NxN)')
    plt.ylabel('Execution Time (seconds, log scale)')
    plt.title('GPU Implementation Time Comparison\nAcross Different Block Sizes and Matrix Dimensions')
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    
    return plt.gcf()


# Create the three main visualizations
speedup_fig = create_speedup_comparison(df_combined)
error_fig = create_error_comparison(df_combined)
time_fig = create_time_comparison(df_combined)

# Save high-resolution versions of the plots
speedup_fig.savefig('gpu_speedup_comparison.png', dpi=300, bbox_inches='tight')
error_fig.savefig('gpu_error_comparison.png', dpi=300, bbox_inches='tight')
time_fig.savefig('gpu_time_comparison.png', dpi=300, bbox_inches='tight')

# Print summary statistics grouped by block size
print("\nSummary Statistics by Implementation and Block Size:")
summary = df_combined.groupby('Block_Config').agg({
    'Basic GPU Speedup': ['mean', 'max'],
    'Shared GPU Speedup': ['mean', 'max'],
    'CuPy GPU Speedup': ['mean', 'max'],
    'Basic GPU Max Diff': 'mean',
    'Shared GPU Max Diff': 'mean',
    'CuPy GPU Max Diff': 'mean'
}).round(3)

print(summary)

# Identify the best performing configurations for each matrix size
print("\nBest Configurations by Matrix Size:")
best_configs = df_combined.loc[df_combined.groupby('Matrix Size')['CuPy GPU Speedup'].idxmax()]
print(best_configs[['Matrix Size', 'Block_Config', 'Basic GPU Speedup', 
                   'Shared GPU Speedup', 'CuPy GPU Speedup']])
```

Recherche 32 x 16 (beim ersten mal lösen der Aufgabe):
Die 32x16 Blockgröße bietet bei der GPU-Matrixmultiplikation die optimale Performance aufgrund mehrerer wichtiger Hardware-Eigenschaften. Die Breite von 32 Threads passt perfekt zu einem Warp (der kleinsten Ausführungseinheit der GPU), was maximale Ausführungseffizienz gewährleistet. Die Höhe von 16 Threads reduziert dabei Bank-Konflikte im Shared Memory und ermöglicht eine bessere Balance zwischen Parallelität und Ressourcennutzung. Mit insgesamt 512 Threads pro Block wird eine ideale Occupancy erreicht, da mehrere Blocks gleichzeitig auf einem Streaming Multiprocessor (SM) aktiv sein können, was bei größeren Konfigurationen wie 32x32 (1024 Threads) nicht möglich wäre.

Die 32er-Breite unterstützt zudem optimales Memory Coalescing, wodurch zusammenhängende Speicherzugriffe effizient gebündelt werden. Der geringere Registerbedarf im Vergleich zu größeren Blockkonfigurationen erlaubt mehr aktive Warps und damit eine bessere Latenz-Verdeckung. Gleichzeitig wird der Shared Memory effizient genutzt, da genügend Platz für mehrere aktive Blocks bleibt. Diese Kombination aus Warp-Ausrichtung, effizienter Speichernutzung und optimaler Thread-Parallelität führt zu der beobachteten überlegenen Performance bei der Matrixmultiplikation.

4: Wie könnte diese Anwendung in einem Producer-Consumer Design konzipiert werden?

PS. Das Producer-Consumer Design ist ein Modell für parallele Prozesse mit zwei Hauptrollen:
- Der Producer erzeugt Daten und legt sie in einen gemeinsamen Puffer.
- Der Consumer nimmt sie von dort und verarbeitet sie. Der Puffer (meist eine Queue) entkoppelt beide Prozesse, sodass sie unabhängig und mit verschiedenen Geschwindigkeiten arbeiten können.
<br><br>

<font color='blue'>

Ein klassisches Beispiel ist ein Videoplayer: Der Producer lädt Videodaten, während der Consumer sie parallel dekodiert und anzeigt.

Producer:
Lädt die komprimierten SVD-Komponenten (U, S, V Matrizen)
Verteilt die Matrizen in sinnvolle Teilaufgaben für die Rekonstruktion
Legt diese in einer Queue ab

Consumer:
Mehrere Consumer für parallele Rekonstruktion
Jeder Consumer nutzt eine der Beschleunigungsmethoden (Numpy, Numba, CuPy etc.)
Rekonstruieren Teilbereiche des Bildes
Ein "Collector" Consumer fügt die Teile wieder zusammen
